```python
import gym
import numpy as np
from gym.envs.toy_text.frozen_lake import generate_random_map

class FrozenLake10x10Env(gym.envs.toy_text.frozen_lake.FrozenLakeEnv):
    def __init__(self):
        desc = generate_random_map(size=10, p=0.75)
        #desc = [[c if c != b'H' else -1 for c in row] for row in desc]
        super().__init__(desc=desc)

def Env10by10():
    while True:
        env = FrozenLake10x10Env()
        desc = env.desc.tolist()
        num_holes = np.sum([row.count(b'H') for row in desc])
        if (num_holes==25):
            return env

env = Env10by10()
env.render()
```

```python
size = 10
p=0.25
n_states = size*size
n_holes = int(n_states * p)

desc = ['S' + 'F'*(size-1)] * (size-1) + ['S' + 'F'*(size-1) + 'G']

# Randomly place holes
hole_locs = np.random.choice(n_states, size=n_holes, replace=False)

for loc in hole_locs:
    row = loc // size
    col = loc % size
    desc[row] = desc[row][:col] + 'H' + desc[row][col+1:]
desc
```

```python
# Set the new reward values
new_rewards = {
    b'H': -1,   # Hole
    b'G': 1,   # Goal
    b'F': 0,   # Frozen
    b'S': 0    # Start
}

# Get the current transition model
transition_model = env.P

# Update the reward values
for state in transition_model:
    for action in transition_model[state]:
        for i, next_state_prob in enumerate(transition_model[state][action]):
            next_state = next_state_prob[1]
            reward = new_rewards.get(env.desc.flat[next_state])
            transition_model[state][action][i] = (next_state_prob[0], next_state, reward, next_state_prob[3])

print(transition_model)
```
